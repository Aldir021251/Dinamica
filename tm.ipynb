```python
import matplotlib.pyplot as plt
import pandas as pd
# primer ejercicio
# Cargar los datos (asegúrate de que el nombre coincida con tu archivo)
df = pd.read_csv("sentandose.csv")

# Renombrar columnas para facilitar el manejo si es necesario
df.columns = ['time', 'ax', 'ay', 'az', 'abs']

# Crear la gráfica
plt.figure(figsize=(12, 6))
plt.plot(df['time'], df['ax'], label='Aceleración X')
plt.plot(df['time'], df['ay'], label='Aceleración Y')
plt.plot(df['time'], df['az'], label='Aceleración Z')

plt.title('Señales de Aceleración Lineal - Sensor en Pecho')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (m/s^2)')
plt.legend()
plt.grid(True)
plt.show()
df                                                                                                                                                                                  
```
